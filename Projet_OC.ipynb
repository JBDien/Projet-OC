```python
# Récupération des codes pays normés sur le site de la World Bank
df = pd.read_html('https://wits.worldbank.org/wits/wits/witshelp/content/codes/country_codes.htm')
WB_pays = df[0].iloc[2:,0:2]
WB_pays
```

```python
# Détermination du nombre de pays présents dans le dataset de la Banque Mondiale
len(WB_pays[1].unique())
```

```python
# Affichage de la liste des pays présents dans le datasaet de la Banque Mondiale sous forme de dataframe avec les colonnes ( Nom du pays, Code du pays)
WB_pays = WB_pays.rename({0 : "WB Country Name", 1 : "WB Country Code"}, axis='columns')
WB_pays
```

```python
# Détermination de la liste commune des pays ente la liste de la Banque Mondiale et celle du dataset à étudier

liste_codes_communs = []
liste_codes_WB_uniq = []
# Itération sur la liste des codes pays
for i in WB_pays["WB Country Code"]:
# Si le code pays dans la liste de la Banque Mondiale apparaît dans la liste Code_pays on le stocke dans une liste liste_codes_communs
# dans le cas contraire on stocke ce code de pays dans liste_codes_WB_uniq
  if i in Code_pays:
    liste_codes_communs.append(i)
  else:
    liste_codes_WB_uniq.append(i)
WB_pays[WB_pays["WB Country Code"].isin(liste_codes_communs)]
```

```python
# Dataframe des pays uniquement présents dans la liste World Bank
WB_pays[WB_pays["WB Country Code"].isin(liste_codes_WB_uniq)]
```

```python
# # Dataframe des pays uniquement présents dans le dataset et pas dans le site de la Banque Mondiale

liste_codes_dataset_uniq=[]

for i in Code_pays:
  if i not in liste_codes_communs:
    liste_codes_dataset_uniq.append(i)

df = Country[['Country Code','Short Name']]
Country_notWB = df[df['Country Code'].isin(liste_codes_dataset_uniq)]
Country_notWB 

```

```python
# suppression des colonnes vides :

def suppr_col_vides(df) :
  cols_vides = [col for col in df.columns if df[col].isnull().all()]
  df=df.drop(cols_vides, axis=1)
  return df
```

```python
# suppression des lignes vides :

def suppr_lignes_vides(df) :
  lignes_vides = [lignes for lignes in df.index if df.index.isnull().all(1)]
  df.drop(lignes_vides, axis=0, inplace=True)
  return df
```

### 4.1.1 Définition des fonctions nécessaires aux calculs des **score de la moyenne et score d'évolution**

```python
# Calcul pour chaque pays de la moyenne de l'indicateur, de la médiane, de l'écart-type et du coefficient d'évolution

def Score_moyenne (df_num):
  df = pd.DataFrame({'Moyenne indicateur': df_num.mean(1), 'Mediane': df_num.median(1), 'Ecart Type' :df_num.std(1), 'Coeff Evol': df_num.std(1)/df_num.mean(1) }).round(2)
  df.sort_values(by = 'Moyenne indicateur', ascending = False).head(20)
  return df
```

```python
# Calcul du taux d'évolution sur la période

def TxEvol(df_num):
  """
  Cette fonction prend comme paramètre un dataframe au format :
  - df_num renvoyé par la fonction Analyse_indicateurs
  - calcule pour chaque pays le taux d'évolution entre la première année non nulle et la dernière année non nulle
  """
  evol = {}
  # on itère sur toutes les lignes du dataframe df_num
  for i in df_num.index :
    # pour calculer un taux d'évolution, on s'assure de ne pas avoir de valeur vide ou nulle dans la série
    s = df_num.loc[i][df_num.loc[i] != 0].dropna()
    # si la série est vide (= de longueur vide) on ajoute au dictionnaire le nom du pays et 0 de taux d'évolution
    if len(s) == 0 :
      evol.update({i:0})
    # dans le cas contraire, on ajoute au dictionnaire le nom du pays et le taux d'évolution 
    else :
      evol.update({i:(s[-1]-s[0])/s[0]*100})
  # création d'un dataframe avec le dictionnaire créé    
  df = pd.DataFrame(list(evol.items()))
  # les noms de pays sont repositionnés en index, nommage de l'index Country Name et de la la colonne calculée 
  df = df.set_index(0)
  df.columns = ['TxEvol']
  df = df.rename_axis(index='Country Name')
  return df
```

```python
# Calcul du taux d'évolution moyen sur la période (CAGR)

def TxMoyEvol(df_num):
  """
  Cette fonction prend comme paramètre un dataframe au format :
  - df_num renvoyé par la fonction Analyse_indicateurs
  - calcule pour chaque pays le taux d'évolution moyen annuel (CAGR) entre la première année non nulle et la dernière année non nulle
  """
  evol = {}
  for i in df_num.index :
    s = df_num.loc[i][df_num.loc[i] != 0].dropna()
    if len(s) == 0 :
      evol.update({i:0})
    else :
      val_fin = s[-1]
      val_init = s[0]
      annee_fin = int(s.index[-1])
      annee_init = int(s.index[-0])
      if annee_fin == annee_init :
        evol.update({i:0})
      else :
        TxMoyEvol= ((val_fin/val_init)**(1/(annee_fin-annee_init))-1)*100
        evol.update({i : TxMoyEvol})
  df = pd.DataFrame(list(evol.items()))
  df = df.set_index(0)
  df.columns = ['TxMoyEvol']
  df = df.rename_axis(index='Country Name')
  return df
```

### 4.1.2 Fusion des dataframes score moyenne et score d'évolution

```python
# Fusion des dataframes Taux d'évolution et taux d'évolution moyen sur la période

def Evol(df_num):
  df1 = TxEvol(df_num)
  df2 = TxMoyEvol(df_num)
  df_evol = df1.merge(df2, how='inner', left_index=True, right_index=True).round(2)
  df_evol = df_evol.sort_values(by = ['TxEvol'], ascending=False)
  return df_evol
```

```python
# Tableau de synthèse des différents scores

def Score(df_num):
  df1 = Score_moyenne(df_num)
  df2 = Evol(df_num)
  Tab_score = df1.merge(df2, how='inner', left_index=True, right_index=True)
  Tab_score = Tab_score.sort_values(by = ['Moyenne indicateur'], ascending = False)
  return Tab_score
```
